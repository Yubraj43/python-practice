```python
#check even and odd numbers is presnt or not
numbers= [10, 23, 45, 66, 78, 89, 90, 0, -5]
for i in range(numbers):
    if i==0:
        print("zero is neither even nor odd")
    elif i<0:
            print("please enter a positive number")
    elif i%2==0:
            print(f"{i} is even number")
    else:
            print(f"{i} is odd number")
```

```python
#reverse a string
text = input("enter a string:")
reversed_text=""
for char in text :
    reversed_text= char + reversed_text
print(f"reversed string is:{reversed_text}")
```

```python
#reverse a string using for loop
text = input("enter a string:")
reversed_text=""
for char in text[::-1]:
    reversed_text+=char
print(f"reversed string is:{reversed_text}")
```

```python
#prime number checker
num = int(input("enter a number:"))
is_prime=True
for i in range(2,num):
    if num%i==0:
        is_prime=False
        break  
print("prime"if is_prime else"not prime")
```

```python
#prime number lsit 
for num in range(2,101):
    is_prime=True
    for i in range(2,num):
        if num%i==0:
            is_prime=False
            break
        if is_prime:
            print(num)
```

```python
#password strength checker
password =input(("enter your password:"))
digits=alpha=uppercase=special=0
for char in password:
    if char.isdigit():
        digits+=1
    elif char.isalpha():
        alpha+=1
        if char.isupper():
            uppercase+=1
    else:
        special+=1
if len(password)>=8 and digits>=1 and alpha>=1 and uppercase>=1 and special>=1:
    print("strong password")
```

```python
#requency counter
text = input("enter a string:")
freq={}
for char in text:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print(freq) 
```

```python
#duplicate finder 
num = [10, 23, 45, 66, 78, 89, 90, 23, 45, 10]
duplicates = []
for n in num:
    if num.count(n)>1 and n not in duplicates:
        duplicates.append(n)
print(duplicates)
```

```python
#matrix addition
a=[[3,4,5,5,6] ,[2,3,4,5,6,]]
b=[[5,6,7,8,9] ,[1,2,3,4,5]]
result =[]
for i in range(len(a)):
    row=[]
    for j in range(len(a[0])):
        row.append(a[i][j]+b[i][j])
    result.append(row)
print(result)
```
