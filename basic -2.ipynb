```python
#creating percentage of  calculation
name= input("Enter your name: ")
Science= float(input("Enter your science marks: "))
nepali= float(input("Enter your nepali marks: "))
enlish= float(input("Enter your english marks: "))
math=float(input("Enter your math marks: "))
social= float(input("Enter your social marks: "))
total_marks= Science + nepali + enlish + math + social
percentage= (total_marks/500)*100

if percentage>=90:
    grade= "A+ grade"
elif percentage>=80:
    grade= "A grade"   
elif percentage>=70:
    grade= "B+ grade"
elif percentage>=60:
    grade= "B grade"
elif percentage>=50:
    grade= "C grade"    
elif percentage>=40:
     grade= "D grade"      
else:
   grade= "F grade"
print("Hello ", name)
print ("Your total marks are: ", total_marks)
print("Your percentage is: ", percentage)
print("Your grade is: ", grade)
```
